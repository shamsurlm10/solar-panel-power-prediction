```python
import os

import numpy as np
import pandas as pd
import matplotlib.image as img
import matplotlib.patches as patch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("./metadata.xlsx")
```

```python
df.head()
```

```python
images=os.listdir('./sol_dataV2 - Copy')
images=[a.split('.')[0] for a in images]
images
```

```python
dic={
    'panel_name': [],
    'image_name': [],
    'lebel': [],
    'time':[],
}
for panel_name in df['Panel Name']:
    if panel_name in images:
        dic['image_name'].append(f'{panel_name}.jpg')
    else:
        dic['image_name'].append(None)
    dic['panel_name'].append(panel_name)
    dic['time'].append(panel_name.split('-')[1])
    dic['lebel'].append(panel_name.split('-')[0])
new_df=pd.DataFrame(data=dic)
```

```python
new_df=new_df.dropna()
```
